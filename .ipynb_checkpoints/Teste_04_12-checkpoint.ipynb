```python
import pandas as pd
```

```python
df =pd.read_csv('dados/resultados/votacao_candidato_munzona_2020_BRASIL.csv', delimiter=';', encoding='iso-8859-1', usecols=['NM_MUNICIPIO','SG_UF','SQ_CANDIDATO','DS_CARGO','DS_SITUACAO_CANDIDATURA','DS_DETALHE_SITUACAO_CAND','SG_PARTIDO','QT_VOTOS_NOMINAIS'])
```

```python
## Agrupando os candidatos com o mesmo SQ
grouped = df.groupby('SQ_CANDIDATO')
```

```python
## Somando os votos dos candidatos
b = grouped.sum()
```

```python
b
```

```python
type(b)
```

```python
##Agrupando o df que somou os votos com o df original
df1 = pd.merge(df,b, on = "SQ_CANDIDATO") 
```

```python
df1
```

```python
## Retirando a coluna com os votos irreais
df1 = df1.drop(columns=['QT_VOTOS_NOMINAIS_x'])
```

```python
df1
```

```python
## Retitrando os candidatos duplicados
df1 = df1.drop_duplicates()
```

```python
df1
```

```python
dfCandidatosZeroVoto = df1.query('QT_VOTOS_NOMINAIS_y == 0 & DS_SITUACAO_CANDIDATURA=="APTO" & DS_DETALHE_SITUACAO_CAND== "DEFERIDO" & DS_CARGO=="Vereador"')
```

```python
dfCandidatosZeroVoto
```

```python
## Gerando dados para os gráficos
```

```python
## Candidatos com zero votos, por partido e municipio

PartidoMun = dfCandidatosZeroVoto.groupby(['NM_MUNICIPIO','SG_UF','SG_PARTIDO'])['SG_PARTIDO'].count()
PartidoMun
```

```python
## Candidatos com zero votos, por partido

Partido = dfCandidatosZeroVoto.groupby(['SG_PARTIDO'])['SG_PARTIDO'].count()
Partido
```

```python
## Candidatos com zero votos, por estado
Estado = dfCandidatosZeroVoto.groupby(['SG_UF'])['SG_PARTIDO'].count()
Estado
```

```python
dfdescr =pd.read_csv('dados/candidatos/consulta_cand_2020_BRASIL.csv', delimiter=';', encoding='iso-8859-1', usecols=['SQ_CANDIDATO','NR_IDADE_DATA_POSSE', 'DS_GENERO', 'DS_GRAU_INSTRUCAO','DS_COR_RACA','DS_OCUPACAO',])
```

```python
dfdescr.head(5)
```

```python
dfCandidatosZeroVotonovo = pd.merge(dfCandidatosZeroVoto,dfdescr, on = "SQ_CANDIDATO")
```

```python
dfCandidatosZeroVotonovo
```

```python
## Candidatos com zero votos, por Grau de instrução
Grau = dfCandidatosZeroVotonovo.groupby(['DS_GRAU_INSTRUCAO'])['DS_GRAU_INSTRUCAO'].count()
Grau
```

```python
## Candidatos com zero votos, Genero
Genero = dfCandidatosZeroVotonovo.groupby(['DS_GENERO'])['DS_GENERO'].count()
Genero
```

```python
## Candidatos com zero votos,Raça
Raça = dfCandidatosZeroVotonovo.groupby(['DS_COR_RACA'])['DS_COR_RACA'].count()
Raça
```

```python
## Candidatos com zero votos por ocupação
Ocupação = dfCandidatosZeroVotonovo.groupby(['DS_OCUPACAO'])['DS_OCUPACAO'].count()
Ocupação
```
